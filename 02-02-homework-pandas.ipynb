11) Среди кого больше доля зарабатывающих много (>50K): среди женатых или холостых мужчин (признак marital-status)? Женатыми считаем тех, у кого marital-status начинается с Married (Married-civ-spouse, Married-spouse-absent или Married-AF-spouse), остальных считаем холостыми.

```python
marr = [' Married-AF-spouse', ' Married-civ-spouse', ' Married-spouse-absent']
marr_count = df[df['marital.status'].isin(marr)][df['income'] == ' >50K'][df['sex'] == ' Male']['age'].count()
notmarr_count = df[~df['marital.status'].isin(marr)][df['income'] == ' >50K'][df['sex'] == ' Male']['age'].count()
if marr_count > notmarr_count:
    print("Женатых больше: {}".format(marr_count))
else:
    print("Неженатых больше: {}".format(notmarr_count))
```
