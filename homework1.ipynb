```python
import copy

class Matrix:
    def __init__(self, m, n):
        self.Matr = []
        self.m = m
        self.n = n
        for i in range(m):
            self.Matr.append([0] * n)
    
    def print(self):
        for i in range(len(self.Matr)):
            print(self.Matr[i])
        return
    
    def get_cell(self, i, j):
        if (i < self.m or j < self.n) and i >= 0 and j >= 0:
            return self.Matr[i][j]
        print("Недопустимый индекс")
        return None
 
    def set_cell(self, i, j, value):
        if (i < self.m or j < self.n) and i >= 0 and j >= 0:
            self.Matr[i][j] = value
        else:
            print("Недопустимый индекс")

    def sum_mat(self, B):
        if self.n == B.n and self.m == B.m:
            C = Matrix(self.m, self.n)
            for i in range(self.m):
                for j in range(self.n):
                    C.set_cell(i,j, self.Matr[i][j] + B.get_cell(i,j))
            return C
        print("Матрицы разной размерности")
        return None
    
    def mult_numb(self,b):
        if isinstance(b, (int, float)):
            C = Matrix(self.m, self.n)
            for i in range(self.m):
                for j in range(self.n):
                    C.set_cell(i,j, (self.Matr[i][j]) * b)
            return C
        
        print("Второй аргумент не число")
        return None
    
    def mult(self,B):
        if self.n == B.m:
            C = Matrix(self.m, B.n)
            for i in range(self.m):
                for j in range(B.n):
                    d = 0
                    for s in range(self.n):
                        d += self.Matr[i][s] * B.get_cell(s,j)
                    C.set_cell(i, j, d)
            return C
        print("Количество столбцов в первой матрице не равно количеству строк во второй ")
    
    def det(self):
        if self.m == self.n:
            d = 1
            eps = 0.000001

            for i in range(self.m):

                if abs(self.Matr[i][i]) < eps:
                    flag = True
                    for k in range(i + 1, self.m):
                        if abs(self.Matr[k][i]) > eps:
                            flag = False
                            a = copy.deepcopy(self.Matr[k])
                            self.Matr[k] = copy.deepcopy(self.Matr[i])
                            self.Matr[i] = copy.deepcopy(a)
                            d *= -1
                            break
                    if flag:
                        return 0
                for j in range(i + 1, self.m):
                    if self.Matr[j][i] != 0:
                        multiplier = self.Matr[j][i] / (self.Matr[i][i])
                        for k in range(i, self.m):
                            self.Matr[j][k] -= multiplier*self.Matr[i][k]

                d *= self.Matr[i][i]
            return d
        print("Матрица не квадратная")
        return None

    def inverse(self):
        D = copy.deepcopy(self)
        if D.det() != 0 and D.det() != None:
            eps = 0.000001
            M = []
            for i in range(self.m):
                M.append(copy.deepcopy(self.Matr[i]))
                for j in range(self.m):
                    M[i].append(0)
                M[i][self.m + i] = 1

            for i in range(self.m):
                if abs(M[i][i]) < eps:
                    for k in range(i + 1, self.m):
                        if abs(M[k][i]) > eps:
                            a = copy.deepcopy(M[k])
                            M[k] = copy.deepcopy(M[i])
                            M[i] = copy.deepcopy(a)
                            break

                if M[i][i] != 1:
                    multiplier = 1 / M[i][i]
                    for k in range(i, len(M[i])):
                        M[i][k] *= multiplier

                for j in range(self.m):
                    if j == i:
                        continue
                    if M[j][i] != 0:
                        multiplier = M[j][i]
                        for k in range(i,  len(M[i])):
                            M[j][k] -= multiplier*M[i][k]
            I = Matrix(self.m, self.m)
            for i in range(self.m):
                for j in range(self.m):
                    I.set_cell(i, j, round(M[i][j + self.m], 3))
            return I

        print("Матрица вырожденная")
        return None

    def transp(self):
        M = Matrix(self.n, self.m)
        for i in range(self.m):
                for j in range(self.n):
                    M.set_cell(j, i, self.Matr[i][j])
        return M
    
    
print("D")
D = Matrix(3,3)
D.set_cell(0,0,1)
D.set_cell(1,1,1)
D.set_cell(2,2,1)
D.set_cell(2,0,5)
D.set_cell(1,0,1.8)
D.set_cell(0,1,6.1)
D.print()
print("invD")
D.inverse().print()
print("detD")
D.det()
```

```python
def head(pair):
    if pair is None:
        return None
    return pair[0]

def tail(pair):
    if pair is None:
        return None
    if len(pair) < 2:
        return None
    return pair[1]

def Seq(*elements):
    def first(x):
        return x[0]
    def other(x):
        return x[1:]

    if len(elements) == 0:
        return None

    if len(elements) == 1:
        return first(elements), None

    if len(elements) == 2:
        return first(elements), other(elements)

    return first(elements), Seq(*other(elements))

def at(pair, index):
    if pair is None:
        return None
    
    if index == 0:
        return  pair[0]
    
    if len(pair) < 2:
        return None
    
    index -= 1
    return at(pair[1], index)

def size(pair):
    if pair is None:
        return None
    
    if len(pair) < 2:
        return 1
    
    return 1 + size(pair[1])

def eq(pair1, pair2):
    if (pair1 is None) and (pair2 is None):
        return True
    
    if pair1[0] != pair2[0]:
        return False
    
    if len(pair1) < 2 and len(pair2) < 2:
        return True
    
    if len(pair1) < 2 or len(pair2) < 2:
        return False    
    return eq(pair1[1], pair2[1])

def tailN(pair, N):
    if pair is None:
        return None
    
    if N == 0:
        return  pair
    
    N -= 1
    return tailN(pair[1], N)

def concat(pair1, pair2):
    if (pair1 is None) and (pair2 is None):
        return None
    
    if (pair1 is None):
        return pair2
    
    if (pair2 is None):
        return pair1
    
    if len(pair1) < 2:
        return pair1[0], pair2
    
    return pair1[0], concat(pair1[1], pair2)

def map_(seq, func):
    seq2 = []
    for el in seq:
        seq2.append(func(el))
    return seq2
#def for_each_indexed 

#def maps( fun = lambda x: x**2, pair):

s = Seq(1,2,3,4,5)
s == (1, (2, (3, (4, (5,)))))
head(s) == 1
tail(s) == (2, (3, (4, (5,))))
at(s, 3) == 4
size((4,)) == 1
eq(s, (1, (2, (3, (4, (5,))))))
tailN(s, 4) 
concat((1,(2,)),(3,(4,(5,)))) == s
for_each(s)
```
